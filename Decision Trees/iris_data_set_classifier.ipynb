```python
from sklearn import datasets

iris = datasets.load_iris()

# print(iris)
X = iris.data[:, [2,3]]
y = iris.target

# print(f'X: {X}')
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion='entropy', max_depth=4, random_state=1)
classifier.fit(X, y)

```

```python
import numpy as np
import matplotlib.pyplot as plt


def entropy(p):
    return - p * np.log2(p) - (1-p) * np.log2(1-p)


x = np.arange(0.0, 1.0, 0.01)
e = [entropy(i) if i != 0 else None for i in x]


plt.xlabel('p(i=1)')
plt.plot(x, e, label='entropy', color='r')
```

```python
from pydotplus.graphviz import graph_from_dot_data
from sklearn.tree import export_graphviz


dot_data = export_graphviz(
    classifier, filled=True, rounded=True,
    class_names=['Setosa', 'Versicolor','Virginica'],
    feature_names=['petal length', 'petal width'],
    out_file=None
)

graph = graph_from_dot_data(dot_data)
graph.write('File.png', format='png')
```
